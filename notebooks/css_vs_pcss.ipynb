```python
from notebook_utils.utils import *
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import wishart
import itertools
from tqdm.notebook import tqdm
import seaborn as sns
```

```python
def obj_css(Sigma, idxs, resid=None):
    p = len(Sigma)
    k = len(idxs)
    if resid is None:
        resid = regress_off(Sigma, idxs)
    return np.log(np.trace(resid)/(p-k))

def obj_sph_pcss(Sigma, idxs, resid=None, det=None):
    p = len(Sigma)
    k = len(idxs)
    if resid is None:
        resid = regress_off(Sigma, idxs)
    if det is None:
        det = np.linalg.det(Sigma[idxs,:][:, idxs])
    return np.log(det)/(p-k) + np.log(np.trace(resid)/(p-k))

def sample_params(p, k, prop=0.1):
    A = np.random.normal(0, 1, size=(p, k))
    D = k*prop/(1-prop) * np.square(np.random.normal(0, 1,  size=p))
    return A, D

def standardize_params(A, D):
    scale = np.sum(np.square(A), axis=1) + D
    A = np.sqrt(1/scale)[:, np.newaxis] * A
    D = 1/scale * D
    return A, D

def get_sample_cov(n, A, D):
    Sigma_hat = wishart.rvs(df=n-1, scale=A.dot(A.T) + np.diag(D))
    return Sigma_hat 
```

```python
p=10
k=4
n=500

idxs_list = [np.array(idxs) for idxs in list(itertools.combinations(range(p), k))]

A, D = sample_params(p, k)
A, D = standardize_params(A, D)
Sigma_hat = 1/n * get_sample_cov(n, A, D)
css_objectives = np.zeros(len(idxs_list))
sph_pcss_objectives = np.zeros(len(idxs_list))
for j, idxs in tqdm(enumerate(idxs_list)):
    resid = regress_off(Sigma_hat, idxs)
    det = np.linalg.det(Sigma_hat[idxs, :][:, idxs])
    css_objectives[j] = obj_css(Sigma_hat, idxs, resid=resid)
    sph_pcss_objectives[j] = obj_sph_pcss(Sigma_hat, idxs, resid=resid, det=det)
```

```python
plt.scatter(css_objectives, sph_pcss_objectives)
plt.xlabel('Log of CSS Objective')
plt.ylabel('Spherical Gaussian PCSS Objective')
#plt.savefig('../figures/css_vs_pcss_' + str(p) + '.pdf')
```
